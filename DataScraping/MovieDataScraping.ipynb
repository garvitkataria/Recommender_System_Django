```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
```

```python
df = pd.read_csv('data/movies.csv')
df.head()
```

```python
movie_titles = pd.read_csv('data/links.csv')
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on='movieId')
df.head()
```

# Scraping data from IMDB 

```python
t = df["title"]
imdbId = df["imdbId"]
g = df["genres"]

list_Images=[]
list_Dates=[]

mydriver = webdriver.Chrome()


for i in range(300):
    print(t[i],g[i])
    print("https://www.imdb.com/title/tt0"+str(imdbId[i]))
#     //*[@id="title-overview-widget"]/div[1]/div[2]/div/div[2]/div[2]/div/a[3]
    finalDate=11111111
    tempImg="http://reelcinemas.ae/Images/Movies/not-found/no-poster.jpg"
    
    baseurl = "https://www.imdb.com/title/tt0"+str(imdbId[i])
    mydriver.get(baseurl)
    
    try:
        date = mydriver.find_element_by_xpath('/html/body/div[2]/div/div[2]/div[5]/div[1]/div[2]/div/div[1]/div[2]/div/div[2]/div[2]/div/a[4]')
        s1=date.text;
        DD = int(s1.split()[0])   
        MM=s1.split()[1]
        M=0
        if(MM=="January"):
            M=1
        elif(MM=="February"):
            M=2
        elif(MM=="March"):
            M=3
        elif(MM=="April"):
            M=4
        elif(MM=="May"):
            M=5
        elif(MM=="June"):
            M=6
        elif(MM=="July"):
            M=7
        elif(MM=="August"):
            M=8
        elif(MM=="September"):
            M=9
        elif(MM=="October"):
            M=10
        elif(MM=="November"):
            M=11
        elif(MM=="December"):
            M=12
        Y = int(s1.split()[2])
        finalDate = DD*1000000+M*10000+Y
    except NoSuchElementException:
        pass
    if(finalDate==11111111):
        print("Error: Field Not Available(Release Of Date)") 
    print("finalDate: ",finalDate) 
    list_Dates.append(finalDate)

    images = mydriver.find_elements_by_tag_name('img')
    print("image:-")
    print(len(images))
    if(len(images)>=8):
        tempImg=images[8].get_attribute('src')
        print(tempImg)
    list_Images.append(tempImg)
```

```python
len(list_Dates)
```

```python
df_date = pd.DataFrame(list_Images, columns=['Image'])
df_image = pd.DataFrame(list_Dates, columns=['Date'])
print(df_date.head())
print(df_image.head())
```

```python
df.head()
```

```python
bigdata = pd.concat([df,df_date, df_image], axis=1)
bigdata.head()
```

```python
out = bigdata.to_json(orient='records')[1:-1].replace('},{', '} {').replace('\\', '')
out[:1000]
```

```python
with open('ScrappedMovieData.json', 'w') as f:
    f.write(out)
```
